```python
# drop unnecesary columns for this project.

# NonStopMiles - introduce others outside factors out of the scope of this project.
# MktFare - shows some very cheap discounts acting as outliers.
# Only an 8% of total flights represent those that are not in the adjacent States in column ItinGeoType, MktGeoType
# Passengers column represent 78% - 1 tickeck bought, above 5 tickets are < 1% & above 15 < 0.1%
# MktCoupons - 1 & 2 represent more than 97% of the total.
# OpCarrierChange - 21% Passengers changed Operating carrier.

drop_columns = ['ItinID', 'MktID', 'OriginAirportID', 'OriginAirportSeqID', 'OriginCityMarketID', 'OriginCountry', 'OriginStateFips', 
                'OriginStateName', 'DestAirportID', 'DestAirportSeqID', 'DestCityMarketID', 
                'DestCountry', 'DestStateFips', 'DestStateName', 'AirportGroup', 'WacGroup', 
                'TkCarrierChange', 'TkCarrierGroup', 'OpCarrierGroup', 'RPCarrier', 'OpCarrier',
                'BulkFare', 'MktDistanceGroup', 'ItinGeoType', 'MktGeoType', 'Unnamed: 41', 'OriginState', 'DestState']
df_flight = data.drop(drop_columns, axis = 1)

df_flight = df_flight[df_flight['MktMilesFlown'] == df_flight['NonStopMiles']]
df_flight = df_flight.drop(['NonStopMiles'], axis = 1)
df_flight = df_flight[df_flight['MktDistance'] == df_flight['MktMilesFlown']]
df_flight = df_flight.drop(['MktDistance'], axis = 1)
df_flight = df_flight[(df_flight['MktMilesFlown'] >= 100) & (df_flight['MktMilesFlown'] <= 2000)]
df_flight = df_flight[df_flight['OpCarrierChange'] == 0]
df_flight = df_flight.drop(['OpCarrierChange'], axis = 1)
df_flight = df_flight[(df_flight['MktFare'] >= 50) & (df_flight['MktFare'] <= 2000)]
df_flight = df_flight[df_flight['Passengers'] <= 4]
df_flight = df_flight[df_flight['MktCoupons'] <= 3]

df_flight
```

```python
def modify (year, quarter):
    if (year == 2022) & (quarter == 1):
        return 5
    else:
        return quarter

quarter_mod = list(map(modify, df_flight['Year'], df_flight['Quarter']))

df_flight['Quarter'] = quarter_mod
```
